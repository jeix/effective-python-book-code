## 8. Robustness and Performance

### 70 Profile Before Optimizing

```python
def insertion_sort(data):
    result = []
    for value in data:
        insert_value(result, value)
    return result
```

```python
def insert_value(array, value):
    for i, existing in enumerate(array):
        if existing > value:
            array.insert(i, value)
            return
    array.append(value)
```

```python
from random import randint
```

```python
max_size = 10**4
data = [randint(0, max_size) for _ in range(max_size)]
test = lambda: insertion_sort(data)
```

```python
from cProfile import Profile
```

```python
profiler = Profile()
profiler.runcall(test)
```

```python
from pstats import Stats
```

```python
from sys import stdout as STDOUT
```

```python
stats = Stats(profiler)
#stats = Stats(profiler, stream=STDOUT)
stats.strip_dirs()
stats.sort_stats('cumulative')
stats.print_stats()
```

```python
from bisect import bisect_left
```

```python
def insert_value(array, value):
    i = bisect_left(array, value)
    array.insert(i, value)
```

```python
profiler = Profile()
profiler.runcall(test)

stats = Stats(profiler, stream=STDOUT)
stats.strip_dirs()
stats.sort_stats('cumulative')
stats.print_stats()
```

```python
def my_utility(a, b):
    c = 1
    for i in range(100):
        c += a * b

def first_func():
    for _ in range(1000):
        my_utility(4, 5)

def second_func():
    for _ in range(10):
        my_utility(1, 3)

def my_program():
    for _ in range(20):
        first_func()
        second_func()
```

```python
profiler = Profile()
profiler.runcall(my_program)

stats = Stats(profiler, stream=STDOUT)
stats.strip_dirs()
stats.sort_stats('cumulative')
stats.print_stats()
```

```python
stats = Stats(profiler, stream=STDOUT)
stats.strip_dirs()
stats.sort_stats('cumulative')
stats.print_callers()
```

> - 파이썬 프로그램을 느리게 하는 원인이 불분명한 경우가 많으므로 프로그램을 최적화하기 전에 프로파일링하는 것이 중요하다.
> - `profile` 모듈 대신 `cProfile` 모듈을 사용하라. `cProfile`이 더 정확한 프로파일링 정보를 제공한다.
> - 함수 호출 트리를 독립적으로 프로파일링하고 싶다면 `Profile` 객체의 `runcall` 메서드를 사용하기만 하면 된다.
> - `Stats` 객체를 사용하면 프로파일링 정보 중에서 프로그램 성능을 이해하기 위해 살펴봐야 할 부분만 선택해 출력할 수 있다.
